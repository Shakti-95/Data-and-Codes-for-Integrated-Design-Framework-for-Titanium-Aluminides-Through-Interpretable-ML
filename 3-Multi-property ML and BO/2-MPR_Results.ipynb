```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_excel('MPR models results_MAE.xlsx', index_col=0)
df1 = pd.read_excel('MPR models results_RMSE.xlsx', index_col=0)
df2 = pd.read_excel('MPR models results_R2.xlsx', index_col=0)
```

```python
norm_df = (df - df.min(0))/(df.max(0) - df.min(0))
norm_df1 = (df1 - df1.min(0))/(df1.max(0) - df1.min(0))
norm_df2 = (df2 - df2.min(0))/(df2.max(0) - df2.min(0))
```

```python
xtl = [r'$log_{10}\sigma_y$', r'$log_{10}\sigma$', r'$log_{10}\delta$', r'$log_{10}Cost$']
```

```python
from textwrap import wrap


figure, ax = plt.subplots(1, 3, figsize=(32, 18))

ax[0] = sb.heatmap(norm_df, cmap='vlag', annot = df, fmt=".3f", linewidths=1, cbar=False, annot_kws={"size":30},
                   square = False, ax=ax[0])
ax[0].tick_params(left=False, bottom=False, pad=15)
ax[0].set_ylabel(r'$Regressors$', fontsize = 40, labelpad=25)
ax[0].set_xlabel('(a) '+r'$MAE$', fontsize = 40, labelpad=25)
ax[0].set_yticklabels(ax[0].get_yticklabels(), rotation=0, style='italic')
ax[0].set_xticklabels(xtl, rotation=70)
for label in (ax[0].get_xticklabels() + ax[0].get_yticklabels()):
    label.set_fontsize(32)

ax[1] = sb.heatmap(norm_df1, cmap='vlag', annot = df1, fmt=".3f", linewidths=1, cbar=False, annot_kws={"size":30},
                    square=False, yticklabels = False, ax=ax[1])
ax[1].tick_params(left=False, bottom=False)
ax[1].set_ylabel('')
ax[1].set_xlabel('(b) '+r'$RMSE$', fontsize = 40, labelpad=25)
ax[1].set_xticklabels(xtl, rotation=70)
for label in (ax[1].get_xticklabels()):
    label.set_fontsize(32)

ax[2] = sb.heatmap(norm_df2, cmap='vlag_r', annot = df2, fmt=".3f", linewidths=1, cbar=False, annot_kws={"size":30},
                    square = False, yticklabels = False, ax=ax[2])
ax[2].tick_params(left=False, bottom=False)
ax[2].set_ylabel('')
ax[2].set_xlabel('(c) '+r'$R^2$', fontsize = 40, labelpad=25)
ax[2].set_xticklabels(xtl, rotation=70)
for label in (ax[2].get_xticklabels()):
    label.set_fontsize(32) 

plt.tight_layout()
plt.show()
```

```python
figure.savefig('Plots/Multi-property models Results.jpg',dpi=600)
```

```python
figure2, ax = plt.subplots(figsize=(25, 25))

ax = sb.heatmap(norm_df2, cmap='vlag_r', annot = df2, fmt=".3f", linewidths=3, cbar=True, annot_kws={"size":30}, square = True)
ax.tick_params(left=False, bottom=False, pad=15)
ax.set_ylabel(r'$Regressors$', fontsize = 40, labelpad=25)
ax.set_xlabel(r'$R^2$', fontsize = 40, labelpad=25)
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, style='italic')
ax.set_xticklabels(xtl, rotation=70)
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(32)
# use matplotlib.colorbar.Colorbar object
cbar = ax.collections[0].colorbar
# here set the labelsize
cbar.ax.tick_params(length=8, labelsize=25, pad = 8)

plt.tight_layout()
plt.show()
```

```python
figure2.savefig('Plots/Multi-property models Results_R2_cbar.jpg',dpi=600)
```

```python
mae_df = pd.concat([df.iloc[0,:],df.iloc[2,:]],axis=1)
rmse_df = pd.concat([df1.iloc[0,:],df1.iloc[2,:]],axis=1)
r2_df = pd.concat([df2.iloc[0,:],df2.iloc[2,:]],axis=1)
```

```python
figure4, ax = plt.subplots(1, 3, figsize=(39, 15))

ax[0] = mae_df.plot(kind="bar", ax=ax[0], colormap='Dark2')
ax[0].tick_params(length=10, width=1.5)
ax[0].set_ylabel(r'$MAE$', fontsize = 40, labelpad=25)
ax[0].set_xlabel('(d)', fontsize = 40, labelpad=25)
ax[0].set_xticklabels(xtl, rotation=70)
ax[0].legend(fontsize=28)
for label in (ax[0].get_xticklabels() + ax[0].get_yticklabels()):
    label.set_fontsize(32)

ax[1] = rmse_df.plot(kind="bar", ax=ax[1], colormap='Dark2')
ax[1].tick_params(length=10, width=1.5)
ax[1].set_ylim([0.0,0.7])
ax[1].set_ylabel(r'$RMSE$', fontsize = 40, labelpad=25)
ax[1].set_xlabel('(e)', fontsize = 40, labelpad=25)
ax[1].set_xticklabels(xtl, rotation=70)
ax[1].legend(fontsize=28)
for label in (ax[1].get_xticklabels() + ax[1].get_yticklabels()):
    label.set_fontsize(32)

ax[2] = r2_df.plot(kind="bar", ax=ax[2], colormap='Dark2')
ax[2].tick_params(length=10, width=1.5)
ax[2].set_ylim([0.0,1.3])
ax[2].set_ylabel(r'$R^2$', fontsize = 40, labelpad=25)
ax[2].set_xlabel('(f)', fontsize = 40, labelpad=25)
ax[2].set_xticklabels(xtl, rotation=70)
ax[2].legend(fontsize=28)
for label in (ax[2].get_xticklabels() + ax[2].get_yticklabels()):
    label.set_fontsize(32)

plt.tight_layout(pad=5)
plt.show()
```
